```python
# How do you simulate two correlated random variables with variance 1 and correlation rho?
%matplotlib inline
n = 1000
mu1, s1 = 0, 1
mu2, s2 = 0, 1
v1 = np.random.normal(mu1, s1, n)
v2 = np.random.normal(mu2, s2, n)
plt.figure()
plt.plot(v1)
plt.plot(v2)
rho = 0.8
w1 = v1
w2 = rho * v1 + math.sqrt(1-rho**2) * v2
plt.figure()
plt.plot(w1)
plt.plot(w2)
plt.figure()
plt.plot(w1, w2 ,'+')
```

```python
# How do you calculate the five number summary of a numpy array?
x = np.random.normal(2.3, 3.4, 1000)
type(x)
try:
    x.describe()  # Wrong, because np array doesn't have a describe function.
except:
    def summary(x):
        from scipy import stats
        tmp = stats.describe(x)
        print("nObs: %.0f" %tmp.nobs)
        print("MinMax: %.3f, %.3f" %(tmp.minmax[0], tmp.minmax[1]))
        print("Mean: %.3f" %tmp.mean)
        print("Variance: %.3f" %tmp.variance)
        print("Skewness: %.3f" %tmp.skewness)
        print("Kurtosis: %.3f" %tmp.kurtosis)
        print("2.5%%: %.3f" %np.percentile(x, 2.5))  # Double %% is the way to escape % in a string
        print("97.5%%: %.3f" %np.percentile(x, 97.5))  # Percentile function takes numbers as percentiles
    summary(x)
    
```

```python
# How do you simulate all the discrete distributions in Python?
# 1. Bernoulli trials
np.random.binomial(1, 0.5, 20)  # (n, p, size)
```

```python
# 2. Binomial distribution
np.random.binomial(5, 0.5, 20)  # (n, p, size)
```

```python
# 3. Poisson distribution
np.random.poisson(3, 20)  # (lambda, size)
```

```python
# 4. Negative binomial
# If one throws a die repeatedly until the third time a “1” appears, 
# then the probability distribution of the number of non-“1”s that 
# appear before the third “1” is a negative binomial distribution.
np.random.negative_binomial(2, 0.5, 100)  # (n, p, size)
```

```python
# 5. Geometric binomial
np.random.geometric(0.3, 15)  # (n, p, size)
```

```python
# 6. Discrete sampling
x = np.array([1, 2, 3, 4])
p = np.array([0.1, 0.2, 0.3, 0.4])
np.random.choice(x, 12, replace=True, p=p)  # The positional argument replace happens before p
```

```python
# How do you do a 1-sample t test?
x = np.random.normal(1, 2, 100)
stats.ttest_1samp(x, 2.0)
```

```python
# How do you do a 2-sample t test?
y = np.random.normal(2, 3, 200)
stats.ttest_ind(x, y)
```

```python
# How do you do a paired 2-sample t test?
x = np.random.normal(1, 2, 100)
y = np.random.normal(1.5, 1.5, 100)
z = x - y
stats.ttest_1samp(z, 0)
```

```python
# How do you do a anova test for multiple groups?
x = np.random.normal(1, 2, 100)
y = np.random.normal(1.5, 1.5, 100)
z = np.random.normal(1.2, 1.8, 100)
stats.f_oneway(x, y, z)
```

```python
# How do you do optimization of a simple function with one input parameter?
# Can you do it with the built-in optimizer?
# Can you do it with gradient descent algorithm?
# Can you do it with Newton Ralphson algorithm?
# Can you do it with Nelder-Mead algorithm?
# Can you do it for a multivariate function?
def f(x):
    return (x - 2) ** 2 + math.sqrt(x)

d0 = 0.000001
(f(2+d0) - f(2-d0))/(2*d0)

x = np.arange(0, 100)/10
plt.plot(x, np.array(list(map(f, x))))
```

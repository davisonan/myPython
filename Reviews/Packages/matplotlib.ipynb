### Background
Matplotlib is the central data visualizaiton tool in the Python ecosystem. Different from R where basic graphs like histograms, basic plots are given in the base package, in Python, all the key plotting functions are in this package. Therefore, it's worth the effort to study this package explicitly.

This is a large package with many many examples. This notebook doesn't aim to exhaust all the examples, but to cover the basic and common graphs I encounted and expect to encounter again in the future. Here we go!

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot([1,2,3,4])
plt.xlabel('Index')
plt.ylabel('some numbers')
plt.title('The first graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 10, 0.2)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
plt.show()
```
